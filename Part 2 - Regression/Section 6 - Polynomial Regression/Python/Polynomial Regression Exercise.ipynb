# Polynomial Regression 

## 1. Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 2. Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## 3. Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()

linear_regressor.fit(X, y)
```

## 4. Training the Polynomial Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regressor = PolynomialFeatures(degree = 4)
X_poly = poly_regressor.fit_transform(X)
poly_linear_regressor = LinearRegression()

poly_linear_regressor.fit(X_poly, y)
X_poly
```

## 5. Visualizing the Linear Regression results

```python
plt.scatter(X, y)
plt.plot(X, linear_regressor.predict(X), 'r')
plt.xlim([0, 11])
plt.title('Level vs Salaries (Linear Regression)')
plt.xlabel('Level')
plt.ylabel('Salaries')
```

## 6. Visualizing the Polynomial Linear Regression results

```python
plt.scatter(X, y)
plt.plot(X, poly_linear_regressor.predict(X_poly), 'r')
plt.xlim([0, 11])
plt.title('Level vs Salaries (Polynomial Regression)')
plt.xlabel('Level')
plt.ylabel('Salaries')
```

## 7. Visualizing the Polynomial Regression results(for higher resolution and smoother curve)

```python
plt.scatter(X, y)
plt.plot(X, poly_linear_regressor.predict(X_poly), 'r')
plt.xlim([0, 11])
plt.title('Level vs Salaries (Polynomial Regression)')
plt.xlabel('Level')
plt.ylabel('Salaries')
```

## 8. Predicting a new result with Linear Regression

```python
linear_regressor.predict([[6.5]])
```

## 9. Predicting a new result with Polynomial Regression

```python
poly_linear_regressor.predict([[1, 6.5, 6.5**2, 6.5**3, 6.5**4]])
```
